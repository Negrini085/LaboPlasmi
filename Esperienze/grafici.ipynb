### Tolgo rumore medio

```python
fig = plt.figure(figsize=(15, 6))

# Creo la griglia per gli output
x = np.linspace(0, 1224, 1224)
y = np.linspace(0, 1024, 1024)
X, Y = np.meshgrid(x, y)


#--------------------------------------------------#
#                 Plasma originale                 #
#--------------------------------------------------#
ax = fig.add_subplot(1, 2, 1)
inte = np.array(Image.open("Esperienza 0/Dati/CAMimages/plasma/plasma001.tif"))
# Faccio plot
cax = ax.pcolormesh(X, Y, inte, cmap='jet')
plt.colorbar(cax); ax.set_aspect('equal')
ax.set_title(f'Immagine di plasma originale', fontsize = 20, pad=20)


#--------------------------------------------------#
#              Calcolo rumore medio                #
#--------------------------------------------------#
rumore_medio = 0
for i in range(1, 81):

    # Scelgo il nome del file da aprire
    if i < 10:
        appo = np.array(Image.open("Esperienza 0/Dati/CAMimages/dark2x2/dark00" + str(i) +".tif"))
    else:
        appo = np.array(Image.open("Esperienza 0/Dati/CAMimages/dark2x2/dark0" + str(i) +".tif"))

    rumore_medio = rumore_medio * (i-1)/i + appo/i

ax = fig.add_subplot(1, 2, 2)

# Faccio plot del rumore medio
cax = ax.pcolormesh(X, Y, rumore_medio, cmap='jet', vmin = 0, vmax = 10)
plt.colorbar(cax); ax.set_aspect('equal')
ax.set_title(f'Rumore medio', fontsize = 20, pad=20)

plt.show()
```
