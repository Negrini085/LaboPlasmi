```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

### Profilo di densit√† ideale e reale

```python
fig = plt.figure(figsize=(15, 6))

x = np.linspace(0, 1, 10000)

#Profilo di densit√† ideale: ho un gradino
ax1 = fig.add_subplot(1, 2, 1)
y_id = np.zeros(10000); y_id[0:6000] = y_id[0:6000] + 1
ax1.plot(x, y_id, color = 'blue', linewidth = 2, label = 'Profilo ideale')
ax1.grid(True); ax1.set_xlabel(r'$r/r_W$', fontsize = 14); ax1.set_ylabel(r'$n/n_0$', fontsize = 14)
ax1.set_title('Profilo di densit√† ideale', fontsize = 18)

#Profilo di densit√† reale: √® continuo, con una decrescita graduale
ax2 = fig.add_subplot(1, 2, 2)
y_real = np.zeros(10000); y_real[0:5000] = 1; y_real[5000:10000] = np.exp(-1000 * (x[5000:10000] - 0.5) * (x[5000:10000] - 0.5) * (x[5000:10000] - 0.5))
ax2.plot(x, y_real, color = 'blue', linewidth = 2, label = 'Profilo reale')
ax2.grid(True); ax2.set_xlabel(r'$r/r_W$', fontsize = 14); ax2.set_ylabel(r'$n/n_0$', fontsize = 14)
ax2.set_title('Profilo di densit√† reale', fontsize = 18)
plt.show()
```

### Velocit√† angolari per rotazione rigida

```python
# Creazione del grafico per le frequenze angolari di rotazione rigida della colonna di plasma
x = np.linspace(0, 1, 1000)
s = -4*x*(x-1)

plt.plot(s, x, color = 'blue', linewidth = 2)
plt.xlabel(r'$s$', fontsize = 12); plt.ylabel(r'$\frac{\omega}{\Omega}$', fontsize = 12)
plt.title('Velocit√† angolari', fontsize = 14)
plt.grid(True)

plt.show()
```
